# Deep Learning - creditcard dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental.preprocessing import Normalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
df = pd.read_csv('data/creditcard.csv')
df.head()
```

## Amount forecast

```python
X = df.drop(columns=['Class', 'Amount'])
y = df['Amount']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
# YOUR CODE HERE
def initialize_model_3():
    
    normalizer = Normalization()
    normalizer.adapt(X_train)
    
    model = models.Sequential(normalizer)
    model.add(layers.Dense(20, activation='relu', input_dim=29))
    model.add(layers.Dropout(rate=0.2))

    
    model.add(layers.Dense(20, activation='relu'))
    model.add(layers.Dropout(rate=0.2))
    
    model.add(layers.Dense(10, activation='relu'))
    model.add(layers.Dropout(rate=0.2))
    
    model.add(layers.Dense(1, activation='linear'))
    
    model.compile(loss='mse', optimizer='adam', metrics=['mae'])
    
    return model
```

```python
es = EarlyStopping(patience=15, restore_best_weights=True)

model_2 = initialize_model_3()

history = model_2.fit(X_train, y_train,
                    validation_split=0.3,
                    batch_size=64, 
                    epochs=100, 
                    verbose=1,
                    callbacks=[es])

res_2 = model_2.evaluate(X_test, y_test, verbose=0)
```

# Fraud detection

```python
X = df.drop(columns=['Class', 'Time'])
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
def initialize_model_4():
    
    normalizer = Normalization()
    normalizer.adapt(X_train)
    
    model = models.Sequential(normalizer)
    
    model.add(layers.Dense(40, activation='relu', input_dim=29))
    model.add(layers.Dropout(rate=0.2))

    model.add(layers.Dense(20, activation='relu'))
    model.add(layers.Dropout(rate=0.2))
    
    model.add(layers.Dense(20, activation='relu'))
    model.add(layers.Dropout(rate=0.2))
    
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    return model
```

```python
es = EarlyStopping(patience=10, restore_best_weights=True)

model = initialize_model_4()

history = model.fit(X_train, y_train,
                    validation_split=0.3,
                    batch_size=64, 
                    epochs=150, 
                    verbose=1,
                    callbacks=[es])

res_3 = model.evaluate(X_test, y_test, verbose=0)
```
